```r
install.packages("babynames")

library(babynames)

str(babynames)

bruno_df <- babynames[babynames$name == "Emma", ]

plot(x = bruno_df$year,
    y = bruno_df$n,
    type = "l",
    main = "Popularity of the name Emma")
```
